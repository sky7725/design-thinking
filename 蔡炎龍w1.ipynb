```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.xkcd()
data = pd.DataFrame([[20,50,10,15],[30,30,20,40]],columns=['A','B','C','D']) # data <- data.frame(A=c(20,30),B=c(50,30),C=c(10,20),D=c(15,40))
plt.figure(figsize=(5,5))
plt.pie(data.iloc[0,:],radius=1,wedgeprops=dict(width=0.3,edgecolor='w'),colors=['r','g','b','y'],labels=data.columns)
plt.pie(data.iloc[1,:],radius=0.7,wedgeprops=dict(width=0.1,edgecolor='w'),colors=['r','g','b','y'])
```

```python
plt.figure()
plt.style.use('ggplot') #四種畫風 [u'dark_background', u'bmh', u'grayscale', u'ggplot', u'fivethirtyeight']
x=np.linspace(-10,10,200)

plt.figure(figsize=(12, 5))
axes = plt.subplot(1,3,1)  #等同 grid.arrange(資料,ncol=多少)
axes.plot(x,np.sin(x))
axes2 = plt.subplot(1,3,2)
axes2.plot(x,np.cos(x))
axes3 = plt.subplot(1,3,3)
axes3.plot(x,x*x,color="black")
axes3.plot(x,-x*x)
```
