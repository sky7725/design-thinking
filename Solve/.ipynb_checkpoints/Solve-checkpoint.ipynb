# 繪圖與解聯立方程式: Solow Model

## 在學習 Solow Model 中，這模型在求解過程相對複雜，我們可以利用電腦幫助我們繪圖與理解

## 給定 Solow Model 基本設定如下

$$ Y_t  = K_t^\alpha N_t^{1-\alpha} $$
$$ K_{t+1} = I_t + (1-\delta) K_t $$
$$ I_t = sY_t   $$
$$ \Downarrow $$
- 我們假設 $N_{t+1}/N_t = 1+n$
$$ k_{t+1} (1+n) = sy_t + (1-\delta) k_t = s k_t^{\alpha} + (1-\delta) k_t $$
$$ \Downarrow $$

$$ k_{t+1}  = \frac{1}{1+n}\big[  s k_t^{\alpha} + (1-\delta) k_t\big] $$

# 我們先來繪圖

```python
# 引入需要的 Package

import numpy as np
import matplotlib.pyplot as plt

# 給定模型係數
s = 0.1
delta = 0.1
n = 0.1
alp = 0.4

k_old = np.linspace(0,1.2)

k_new = (1/(1+n)) * ( s*(k_old**alp) + (1-delta) * k_old ) 
```

```python
plt.plot(k_old, k_old) # 45 度線
plt.plot(k_old, k_new) # 每人資本累積方程式
```

# Case: 儲蓄率上升，由 $s = 0.1$ 變成 $s=0.2$

```python
s_new = 0.2

k_new2 = (1/(1+n)) * ( s_new*(k_old**alp) + (1-delta) * k_old ) 

plt.plot(k_old, k_old) # 45 度線
plt.plot(k_old, k_new) # 每人資本累積方程式, s = 0.1
plt.plot(k_old, k_new2, 'k') # 每人資本累積方程式. s = 0.2
```

# 計算 Steady State

## 給定 Solow Model Steady State, 我們有

$$ y  = k^\alpha $$
$$ k(1+n) =sy + (1-\delta)k = sk^\alpha + (1-\delta)k $$
$$ i = s y   $$

事實找 Steady-State 意思是，我們希望找出 $k$ 使得

$$ k(1+n)  - ( sk^\alpha + (1-\delta)k ) = 0 $$

我們可以使用 fsolve

在這之前，我們先介紹 Python 下的一個重要觀念，函數 def

# 當你想要有個工具，替你計算 $x^2 -2x+1$

```python
def f(x):
    return x**2 - 2*x + 1


f(4)
```

上面中的 def 是 Python 中可以用來幫助你自訂函數的工具

```python
n = 0.1
delta = 0.1
s = 0.4
alp = 0.5

def k_ss(k):
    return k*(1+n) - (s*(k**alp) + (1-delta)*k )
    
# 我們希望找到 k 使得 k_ss = 0
```

```python
from scipy.optimize import fsolve
```

```python
fsolve(k_ss, 1)
```

# 計算 Steady State: Multiple Variables

## 給定 Solow Model Steady State, 我們有

$$ y  = k^\alpha $$
$$ k(1+n) =sy + (1-\delta)k  $$
$$ i = s y   $$


事實找 Steady-State 意思是，我們希望找出 $k, y, i$ 使得

$$ y  - k^\alpha =0$$
$$ k(1+n) -sy + (1-\delta)k =0 $$
$$ i - s y  =0 $$

現在有三個方程式，我們依舊可以使用 fsolve

```python
def find_ss(x):
    y = x[0]
    k = x[1]
    i = x[2]
    eq0 = y - k**alp
    eq1 = k*(1+n)- (s*y + (1-delta)*k )
    eq2 = i - s*y
    return eq0, eq1, eq2
```

```python
fsolve(find_ss, [1,1,1])
```

- 解出來結果是，$y^* = 2, k^* = 4, i^* = 0.8$

$$\text{t+1 期: } S = p\cdot N\cdot R_d $$
$$\text{t+2 期: }  (N-S)R_2 = (1-p)\cdot N\cdot R_d^2 $$

我們想要知道銀行利率 $R_d$ 與保留之現金 $S$ 各為多少

# Quadratic Equation

$$ (1-p)R_d^2 + p\cdot R_2 \cdot R_d - R_2 = 0$$

- 假設我們想解 $x^2 - 7x+12$

```python
from numpy import roots
```
